```python

# Generate 1000 random numbers uniformly distributed between [0, 1]
random_numbers = np.random.uniform(0, 1, 1000)

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(random_numbers, bins=100, color='blue', alpha=0.7)

# Labeling the axes and the title
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Histogram of Random Numbers from Uniform Distribution', fontsize=16)

# Save the figure as a PDF
plt.savefig('uniform_distribution_histogram.pdf')

# Show the plot
plt.show()
```
