```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1000

x = np.random.randn(n)

n = 1000

x = np.random.randn(n)

width = 0.5
histmin = np.floor(min(x))
histmax = np.ceil(max(x))+width

bins = np.arange(histmin,histmax,width)
plt.hist(x,bins=bins,alpha=0.5,edgecolor="black")

plt.xlabel('x')
plt.ylabel('N per bin')
plt.title('Histogram of 1000 Random Numbers')

plt.savefig("journal_9.png",bbox_inches="tight",facecolor="white")
plt.show
```
