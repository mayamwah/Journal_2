```python
# Set the seed for reproducibility
np.random.seed(42)

# Parameters for the geometric distribution
p = 0.5  # Probability of success on each trial

# Generate 1000 random numbers from the geometric distribution
random_numbers = np.random.geometric(p, 1000)

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(random_numbers, bins=100, color='blue', alpha=0.7)

# Labeling the axes and the title
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Histogram of Random Numbers from Geometric Distribution', fontsize=16)

# Save the figure as a PDF
plt.savefig('geometric_distribution_histogram.pdf')

# Show the plot
plt.show()
```
