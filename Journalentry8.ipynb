```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def exp_function(x):
    return np.exp(x)
y = exp_function(x)
```

```python
plt.plot(x,y)
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
plt.show()
plt.savefig("journal_entry7.png",bbox_inches="tight",facecolor="white")
```
